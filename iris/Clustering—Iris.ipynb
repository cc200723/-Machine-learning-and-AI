## 聚类分析

```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
#对分类进行处理
def iris_type(s):
    it = {b'Iris-setosa':0,b'Iris-versicolor':1,b'Iris-virginica':2}
    return it[s]
```

```python
#读入数据
path = u'D:/Desktop/iris.data'#数据文件路径
data = np.loadtxt(path,dtype=float,delimiter=',',converters={4:iris_type})
```

```python
#对鸢尾花的花萼长度以及花萼宽度进行聚类分析
x,y = np.split(data,(4,),axis=1)#对数据进行拆分处理
x = x[:,:2]#选取前两列进行分析
clf = KMeans(n_clusters=3)#用KMeans方法将数据聚成三类
model = clf.fit(x)#对聚类后数据进行拟合
label_pred = model.labels_#获取聚类标签
x0 = x[label_pred==0]
x1 = x[label_pred==1]
x2 = x[label_pred==2]
#用红色的“o”表示第一类聚类
plt.scatter(x0[:,0],x0[:,1],c="red",marker='o',label='label0')
#用绿色的“*”表示第二类聚类
plt.scatter(x1[:,0],x1[:,1],c="green",marker='*',label='label1')
#用蓝色的“+”表示第一类聚类
plt.scatter(x2[:,0],x2[:,1],c="blue",marker='+',label='label2')
mpl.rcParams['font.sans-serif'] = ['KaiTi']
plt.xlabel('花萼长度')
plt.ylabel('花萼宽度')
plt.legend(loc=2)
plt.show()
```

```python
#对鸢尾花的花瓣长度以及花瓣宽度进行聚类分析
x,y = np.split(data,(4,),axis=1)#对数据进行拆分处理
x = x[:,[2,3]]#选取后两列进行分析
clf = KMeans(n_clusters=3)#用KMeans方法将数据聚成三类
model = clf.fit(x)#对聚类后数据进行拟合
label_pred = model.labels_#获取聚类标签
x0 = x[label_pred==0]
x1 = x[label_pred==1]
x2 = x[label_pred==2]
#用红色的“o”表示第一类聚类
plt.scatter(x0[:,0],x0[:,1],c="red",marker='o',label='label0')
#用绿色的“*”表示第二类聚类
plt.scatter(x1[:,0],x1[:,1],c="green",marker='*',label='label1')
#用蓝色的“+”表示第一类聚类
plt.scatter(x2[:,0],x2[:,1],c="blue",marker='+',label='label2')
mpl.rcParams['font.sans-serif'] = ['KaiTi']
plt.xlabel('花瓣长度')
plt.ylabel('花瓣宽度')
plt.legend(loc=2)
plt.show()
```
