### 数据读取


```python
# 读取数据
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# 读取数据（路径已按题目要求配置）
train_path = "/bohr/Penguin-dh1y/v5/train.csv" 
test_no_label_path = "/bohr/Penguin-dh1y/v5/test_no_label.csv"

train_df = pd.read_csv(train_path)
test_df = pd.read_csv(test_no_label_path)
```

### 数据预处理
对训练集中的 **数值特征** 进行 **归一化**/**标准化**以及**对非数值特征进行编码**，然后将处理后的特征输出为<font color = "red">**`X.csv`**</font>文件（只需要保存特征，标签数据不要保存）。    
<font color = "red">本文件将会用于最终评分，请务必确保内容格式与名称正常</font>

```python
# 删除含有缺失值的数据
train_clean = train_df.dropna().reset_index(drop=True)
print(f"删除缺失值后训练集样本数：{len(train_clean)}")
```

```python
# 标准化/归一化
from sklearn.preprocessing import StandardScaler, LabelEncoder

# 分离特征和标签
X_train_raw = train_clean.drop(columns=['Species'])
y_train = train_clean['Species']

# 2. 标签编码（仅对训练集标签 fit_transform）
label_encoder = LabelEncoder()
y_train_enc = label_encoder.fit_transform(y_train)
print(f"标签编码器已拟合，包含类别: {list(label_encoder.classes_)}")

# 自动区分数值型和类别型特征
num_cols = X_train_raw.select_dtypes(include=['float64', 'int64']).columns.tolist()
cat_cols = X_train_raw.select_dtypes(include=['object']).columns.tolist()
print(f"数值特征: {num_cols}\n类别特征: {cat_cols}")
```

```python
# 对非数值特征进行编码
# 标准化数值特征
# 3. 标准化数值特征（满足5分要求）
scaler = StandardScaler()
X_train_num_scaled = scaler.fit_transform(X_train_raw[num_cols])
X_train_num_df = pd.DataFrame(X_train_num_scaled, columns=num_cols)

# 4. 独热编码类别特征
X_train_cat_dummies = pd.get_dummies(X_train_raw[cat_cols], drop_first=False).astype(int)
```

```python
# 保存处理后的特征数据到 "X.csv" 文件
# 保存特征数据（不含标签）
X_train_processed = pd.concat([X_train_num_df, X_train_cat_dummies], axis=1)
X_train_processed.to_csv("X.csv", index=False)
print(" X.csv 已保存")
```

### 模型构建
使用训练数据构建机器学习分类模型

```python
# 搭建机器学习模型
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# 使用随机森林（对结构化小样本数据表现稳定，易突破90%）
model = RandomForestClassifier(n_estimators=300, max_depth=10, random_state=42, n_jobs=-1)
model.fit(X_train_processed, y_train_enc)
```
